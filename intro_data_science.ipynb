## 3)  Analyzing data

The table and the data seesm to imple a negative correlation. The way to determine if this is a singificant relationship is to use a t-test.

The most common t-test is the Welch Allen t-test, which only works if the distribution is normal.
Is the distribution normal?

```python
hour_dist= turnstile_df["ENTRIESn_hourly"].hist()
plt.suptitle("Entries per hour across all stations")
plt.xlabel("Entries per hour")
plt.ylabel("Number of occurrences")
```

As we can see above, the hourly distribution of riderships is not a nomral distirbution

```python
not_raining = turnstile_df[turnstile_df["rain"] == 0]
not_raining_hourly=not_raining["ENTRIESn_hourly"]
raining = turnstile_df[turnstile_df["rain"] == 1]
raining_hourly = raining["ENTRIESn_hourly"]

not_raining_mean = not_raining_hourly.mean()
raining_mean = raining_hourly.mean()

(u, pvalue) = scipy.stats.mannwhitneyu(not_raining_hourly,raining_hourly)
print ("mean entries per hour when not raining: %s" % (not_raining_mean))
print ("mean entries per hour when raining: %s" % (raining_mean))
print ("doubles sided p-value of test statistic: %s" % (pvalue * 2, ))
print ("u-value of test statistic: %s" % (u ))
```

There is a problem with the p-value in the current version [scipy](https://github.com/scipy/scipy/issues/4386)
So, from the u-value, we will compute the pvalue

```python
# below code is taken from thi stackoverfloe post http://stats.stackexchange.com/questions/116315/problem-with-mann-whitney-u-test-in-scipy
m_u = len(not_raining_hourly)*len(raining_hourly)/2

sigma_u = np.sqrt(len(not_raining_hourly)*len(raining_hourly)*(len(not_raining_hourly)+len(raining_hourly)+1)/12)

z = (u - m_u)/sigma_u
pval = 2*scipy.stats.norm.cdf(z)
print ("double sided p-value of test statistic: %s" % (pval ))
```

The p-value is less than 0.05. So, we can say that the null hypothesis (that there is no relationship between rain and ridership). There is a significant relationship since it is below the ussual p-vlue of 0.05

## 4) Model for predictring hourly ridership

Looking at the charts above, we see that the hour,day, raining are possible indidcators.

Other possible indicator could be how much precipitation is there ("precipi"), aveage temperature of location ('meantempi')


```python
def normalize_features(array):
   """
   Normalize the features in the data set.
   """
   array_normalized = (array-array.mean())/array.std()
   mu = array.mean()
   sigma = array.std()

   return array_normalized, mu, sigma

def compute_cost(features, values, theta):
    """
    Compute the cost function given a set of features / values, 
    and the values for our thetas.
    """
    
    m = len(values)
    sum_of_square_errors = np.square(np.dot(features, theta) - values).sum()
    cost = sum_of_square_errors / (2*m)

    return cost

def gradient_descent(features, values, theta, alpha, num_iterations):
    """
    Perform gradient descent given a data set with an arbitrary number of features.
    """
    m = len(values)
    cost_history = []

    for i in range(num_iterations):
        # your code here
        predicted=np.dot(features, theta)
        theta= theta + alpha/m*np.dot((values -predicted ),features)
        cost=compute_cost(features, values, theta)
        cost_history.append(cost)
    return theta, pandas.Series(cost_history)

def predictions(dataframe,alpha=0.1,num_iterations=50):
    
    # Select Features rain,precipi,Hour,meantemp
    features = dataframe[['rain', 'precipi', 'hour', 'meantempi','day_week']]
    
    # Add UNIT to features using dummy variables
    dummy_units = pandas.get_dummies(dataframe['UNIT'], prefix='unit')
    features = features.join(dummy_units)
    
    # Values
    values = dataframe['ENTRIESn_hourly']
    m = len(values)

    features, mu, sigma = normalize_features(features)
    features['ones'] = np.ones(m) # Add a column of 1s (y intercept)
    
    # Convert features and values to numpy arrays
    features_array = np.array(features)
    values_array = np.array(values)

   
    # Initialize theta, perform gradient descent
    theta_gradient_descent = np.zeros(len(features.columns))
    theta, cost_history = gradient_descent(features_array, 
                                                            values_array, 
                                                            theta_gradient_descent, 
                                                            alpha, 
                                                            num_iterations)
    
    plot = None
    plot = plot_cost_history(alpha, cost_history)
    
    predictions = np.dot(features_array, theta)
    return predictions, plot,theta


def plot_cost_history(alpha, cost_history):
   """This function is for viewing the plot of your cost history.
   """
   cost_df = pandas.DataFrame({
      'Cost_History': cost_history,
      'Iteration': range(len(cost_history))
   })
   return ggplot(cost_df, aes('Iteration', 'Cost_History')) + \
      geom_point() + ggtitle('Cost History for alpha = %.3f' % alpha )


def plot_residuals(turnstile_weather, predictions):
    
    plt.figure()
    (turnstile_weather - predictions).hist()
    return plt

def compute_r_squared(data, predictions):
    
    data_mean= np.mean(data)
    totals= np.sum((data-data_mean)**2)
    residuals= np.sum((data-predictions)**2)
    
    r_squared= 1 - residuals/totals
    
    # your code here
    
    return r_squared
```
